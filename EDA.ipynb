```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/User/OneDrive/Desktop/DA Intern @ ApexPlanet/Task - 2/Walmart_Sales.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df['Month'] = df['Date'].dt.month_name()
df['Year']  = df['Date'].dt.year
df['Week']  = df['Date'].dt.isocalendar().week
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Weekly_Sales'], bins=30, kde=True)
plt.title("Weekly Sales Distribution")
plt.show()
```

```python
sns.countplot(data=df, x='Store')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=df, x='Holiday_Flag', y='Weekly_Sales')
plt.title("Holiday vs Normal Week Sales")
plt.show()
```

```python
plt.figure(figsize=(12,6))
store_sales = df.groupby('Store')['Weekly_Sales'].sum().reset_index()

sns.barplot(data=store_sales, x='Store', y='Weekly_Sales', palette='viridis')
plt.title('Total Sales by Store')
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
