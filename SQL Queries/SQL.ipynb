```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
# Load CSV
df = pd.read_csv("C:/Users/User/OneDrive/Desktop/DA Intern @ ApexPlanet/Task - 2/Walmart_Sales.csv")

# Show first few rows
df.head()
```

```python
# Convert date from DD-MM-YYYY format
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")

# Additional useful columns
df['Month'] = df['Date'].dt.month
df['Year']  = df['Date'].dt.year
df['Week']  = df['Date'].dt.isocalendar().week

df.head()
```

```python
# Create a new connection (in-memory database)
conn = sqlite3.connect(":memory:")

# Load dataframe into SQL table
df.to_sql("walmart", conn, index=False, if_exists='replace')

print("SQL table 'walmart' created successfully!")
```

```python
query1 = """
SELECT Store, SUM(Weekly_Sales) AS Total_Sales
FROM walmart
GROUP BY Store
ORDER BY Total_Sales DESC;
"""

pd.read_sql_query(query1, conn)
```

```python
query2 = """
SELECT strftime('%m', Date) AS Month,
       AVG(Weekly_Sales) AS Avg_Sales
FROM walmart
GROUP BY Month
ORDER BY Avg_Sales DESC;
"""

pd.read_sql_query(query2, conn)
```

```python
query3 = """
SELECT Holiday_Flag,
       AVG(Weekly_Sales) AS Avg_Sales
FROM walmart
GROUP BY Holiday_Flag;
"""

pd.read_sql_query(query3, conn)
```

```python
query4 = """
SELECT Date, Store, Weekly_Sales
FROM walmart
ORDER BY Weekly_Sales DESC
LIMIT 10;
"""

pd.read_sql_query(query4, conn)
```

```python
query5 = """
SELECT Fuel_Price,
       AVG(Weekly_Sales) AS Avg_Sales
FROM walmart
GROUP BY Fuel_Price
ORDER BY Fuel_Price;
"""

pd.read_sql_query(query5, conn)
```

```python
query6 = """
SELECT strftime('%Y', Date) AS Year,
       SUM(Weekly_Sales) AS Total_Sales
FROM walmart
GROUP BY Year
ORDER BY Year;
"""

pd.read_sql_query(query6, conn)
```

```python
query7 = """
SELECT CPI,
       AVG(Weekly_Sales) AS Avg_Sales
FROM walmart
GROUP BY CPI
ORDER BY CPI;
"""

pd.read_sql_query(query7, conn)
```

```python
query8 = """
SELECT Store, strftime('%Y', Date) AS Year,
       SUM(Weekly_Sales) AS Total_Sales
FROM walmart
GROUP BY Store, Year
ORDER BY Store, Year;
"""

pd.read_sql_query(query8, conn)
```
