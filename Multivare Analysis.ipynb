```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/User/OneDrive/Desktop/DA Intern @ ApexPlanet/Task - 2/Walmart_Sales.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df['Month'] = df['Date'].dt.month_name()
df['Year']  = df['Date'].dt.year
df['Week']  = df['Date'].dt.isocalendar().week
```

```python
numeric_df = df[['Weekly_Sales','Temperature','Fuel_Price','CPI','Unemployment']]

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
df.groupby('Date')['Weekly_Sales'].sum().plot(figsize=(12,5))
plt.title("Weekly Sales Trend Over Time")
plt.show()
```

```python
sns.scatterplot(data=df, x='Temperature', y='Weekly_Sales')
```
